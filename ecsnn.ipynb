```python
import torch

from models import cnn, ecnn, csnn, ecsnn
from datasets import mnist, kmnist, fashionmnist
from pipelines import train

from models import cn
from utils import event_conversion
```

```python
device = torch.device("cuda:0") if torch.cuda.is_available() else torch.device("cpu")
```

### Hyperparameters

```python
IN_CHANNELS = 1
OUT_CHANNELS = 10
HIDDEN_CHANNELS = 32
KERNEL_SIZE = 5

HIDDEN_LAYERS = 4

CYCLIC_GROUP_ORDER = 4

EVENT_CONVERSION = lambda x : event_conversion.random_conversion(x, 8)
# EVENT_CONVERSION = lambda x : event_conversion.trivial_conversion(x, 8)
```

### Models

```python
models = {
    "cnn": lambda : 
        cnn.CNN(
            in_channels=IN_CHANNELS,
            out_channels=OUT_CHANNELS,
            kernel_size=KERNEL_SIZE,
            num_hidden=HIDDEN_LAYERS,
            hidden_channels=HIDDEN_CHANNELS
        ),
    "ecnn": lambda : 
        ecnn.GroupEquivariantCNN(
            group=cn.CyclicGroup(4).to(device),
            in_channels=IN_CHANNELS,
            out_channels=OUT_CHANNELS,
            kernel_size=KERNEL_SIZE,
            num_hidden=HIDDEN_LAYERS,
            hidden_channels=HIDDEN_CHANNELS // 2
        ),
    "csnn": lambda :
        csnn.CSNN(
            in_channels=IN_CHANNELS,
            out_channels=OUT_CHANNELS,
            kernel_size=KERNEL_SIZE,
            num_hidden=HIDDEN_LAYERS,
            hidden_channels=HIDDEN_CHANNELS,
            event_converter=EVENT_CONVERSION
        ),
    "ecsnn": lambda : 
        ecsnn.ECSNN(
            group=cn.CyclicGroup(4).to(device),
            in_channels=IN_CHANNELS,
            out_channels=OUT_CHANNELS,
            kernel_size=KERNEL_SIZE,
            num_hidden=HIDDEN_LAYERS,
            hidden_channels=HIDDEN_CHANNELS // 2,
            event_converter=EVENT_CONVERSION
        )
}
```

### Datasets

```python
datasets = {
    "mnist": mnist.get_RotMNIST(dataset_path="./data"),
    "kmnist": kmnist.get_RotKMNIST(dataset_path="./data"),
    "fashionmnist": fashionmnist.get_RotFashionMNIST(dataset_path="./data")
}
```

### Training

```python
EPOCHS = 16
OPTIMIZER_HPARAMS = {
    "lr": 1e-4,
    "weight_decay": 1e-6
    }
```

```python
with torch.device(device):
    for dataset_name, dataset in datasets.items():
        for model_name, model_generator in models.items():
            _, res = train.train_model(
                model=model_generator(),
                train_loader=dataset[0],
                test_loader=dataset[1],
                save_name=f"./artifacts/{dataset_name}/{model_name}",
                epochs=EPOCHS,
                optimizer_hparams=OPTIMIZER_HPARAMS
            )

            print(model_name, res)
```
